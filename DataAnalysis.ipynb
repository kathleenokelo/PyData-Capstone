#### Top 5 most Popular Categories

```python
final_df['Category'].value_counts()
```

```python
final_df['Category']=final_df['Category'].apply(lambda x: x.lower())

final_df['Category'].value_counts()
```

```python
final_df['Category'].nunique()
```

```python
import re

pattern = r'^".*"$'

def Replace(x):
    match = re.findall(pattern,x)
    if match:
        match = match[0].replace('"','')
        return match
    else:
        return x
    


final_df['Category'] = final_df['Category'].apply(Replace)
final_df['Category'].value_counts()



```

```python
final_df['Category'].nunique()
```

```python
popular_categories= final_df.groupby('Category')[['Score']].sum().reset_index().iloc[:5].sort_values(by='Score',ascending=False)
popular_categories
```

```python
top_reactions.head()
```

```python
# Define the reaction types for each group
group1_reactions = ['hate', 'super love', 'want', 'scared', 'indifferent', 'dislike', 'like']
group2_reactions = ['interested', 'peeking', 'adore', 'heart', 'love']

# Filter data for each group
x1 = top_reactions[top_reactions['Reaction Type'].isin(group1_reactions)]
x2 = top_reactions[top_reactions['Reaction Type'].isin(group2_reactions)]

# Create subplots
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 8))

# Plot for the first group of reactions
sns.barplot(x='Reaction Type', y='Counts', hue='Content Type', data=x1, ax=ax[0],
            dodge=True, edgecolor='black', linewidth=0.2)
ax[0].set_title('Least Common Reactions Per Content Type')
ax[0].set_xlabel('Reaction Type')
ax[0].set_ylabel('Counts')
ax[0].legend(title='Content Type')
ax[0].grid(axis='y')

# Plot for the second group of reactions
sns.barplot(x='Reaction Type', y='Counts', hue='Content Type', data=x2, ax=ax[1],
            dodge=True, edgecolor='black', linewidth=0.2)
ax[1].set_title('Most Common Reactions Per Content Type')
ax[1].set_xlabel('Reaction Type')
ax[1].set_ylabel('Counts')
ax[1].legend(title='Content Type')
ax[1].grid(axis='y')

plt.tight_layout()
plt.savefig('common_reactions.png')
plt.show()
```

#### Correlation between the type of content and the number of reactions

```python
# Create a pivot table to count the number of reactions for each type of content
pivot_table = pd.pivot_table(final_df, index='Content Type', columns='Reaction Type', values='Score', aggfunc='count')

# Fill missing values with 0
pivot_table = pivot_table.fillna(0)

# Normalize the heatmap by the total number of reactions for each content type
pivot_table = pivot_table.apply(lambda x: x / x.sum(), axis=1)

# Create a heatmap
plt.figure(figsize=(18, 10))
sns.heatmap(pivot_table, annot=True, cmap='coolwarm', square=True, fmt='.3f')
plt.title('Heatmap of Reactions by Content Type')
plt.xlabel('Reaction Type')
plt.ylabel('Content Type')
plt.savefig('heatmap_of_reactions.png')
plt.show()
```

If reactions can be categorized as positive or negative, what is the overall sentiment distribution of reactions?


```python
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

final_df['Month'] = pd.Categorical(final_df['Month'], categories=month_order, ordered=True)

sentiment_counts_over_time = final_df.groupby(['Month', 'Sentiment']).size().unstack('Sentiment')
sentiment_counts_over_time
```

```python


fig,ax = plt.subplots(figsize=(10, 8))
sns.lineplot(data=sentiment_counts_over_time)
ax.set_title('Sentiment Distribution of Reactions Over Time')
ax.set_xlabel('Month')
ax.set_ylabel('Count')
ax.set_xlim('January','December')
ax.grid(True)
ax.locator_params(axis='x',nbins=12)
ax.locator_params(axis='y',nbins=10)
ax.tick_params(axis='x',labelrotation=45)
plt.legend(title='Sentiment',loc='center right')

plt.savefig('Sentiment Distribution.png')
plt.show()
```
